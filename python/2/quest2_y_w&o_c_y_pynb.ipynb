```python
#Quest


# 조건
# 텍스트 전처리를 수행한다.
# - Avengers.txt 파일을 불러온다.
# - 모든 문자는 소문자로 변환한다.
# - 모든 기호를 제거한다.
# - 단어 단위로 분리한다.


# 출력결과
# I am hungry very much
# -> ['i', 'am', 'hungry', 'very', 'much']

'''
Avengers.txt 파일을 불러온다.
1. 파일 불러오기
페들릿 안내 링크 보고 확인
2. 모든 문자를 소문자로 변환한다.
문자열을 입력 받았다고 생각하고, 임시 문자열 입력 변수 지정 후 기능 작업
3. 기호 제거 기능

4. 모든 단어 단위로 분리한다.

5. 단어 별 빈도수 저장
 ->

6.

'''

#임시변수
# input_txt = ('/content/drive/MyDrive/아이펠/Avengers.txt')
# input_txt = 'test word good Good GOOD data. hi! '

from google.colab import drive

FILE_PATH = "/content/drive/MyDrive/아이펠/Avengers.txt"

def put_in_file(file_path):
  drive.mount('/content/drive')
  #파일 열기
  file = open("/content/drive/MyDrive/아이펠/Avengers.txt",'r')
  result_txt = ''

  while True:
    line = file.readline()
    result_txt += line
    if not line: break
  file.close()
  return result_txt


input_txt = put_in_file(FILE_PATH)


# 모든 문자를 소문자로 변환한다.
# print(input_txt.lower())
lowered_txt = input_txt.lower()

#특수문자를 제거
# print(input_txt.replace('.','').replace("'",""))

replaced_txt = lowered_txt.replace('.','').replace("'","").replace("!","").replace("?","").replace("\n","").replace("%","").replace("-","")
#replaced_txt = lowered_txt = r'[^\w\s]'

# 단어별로 구분합니다.
# print(replaced_txt.split(" "))
word_list = replaced_txt.split(" ")

# 각 단어 별 빈도수를 딕셔너리 형태로 저장한다.
# 저장에 필요한 딕셔너리를 선언합니다.
word_frequency = {}


for word in word_list:
  if word in word_frequency:
    word_count = word_frequency[word]
    #print(f"word: {word} , count : {word_count+1}")
    word_frequency[word] = word_count+1
  else:
    word_frequency.setdefault(word,1)

# 내림차순 정렬 해서 넣기
vocab = sorted(word_frequency.items(), key=lambda x:x[1], reverse = False)
index = 0
for voca in vocab:
  voca = index
  index += 1

# print(vocab)

input_from_user = input("입력해주세요: ")

frequency_list = []
for text in input_from_user:
  if text in word_frequency:
    frequency_list.append(word_frequency[text])

print(frequency_list)


#print(sorted(word_frequency.items(), key=lambda x:x[1], reverse=True))

#print('we' in word_frequency)
#print(word_frequency['we'])

# 동작확인
# test_var = "good"

# word_frequency.setdefault("good",1)

# print(word_frequency["good"])
# print(word_frequency[test_var])
# #check
# print("good" in word_frequency)

# print(f"test :  {word_frequency.setdefault('good',1)}")
```
