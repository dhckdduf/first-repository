```python
# âœ… Google Drive ë§ˆìš´íŠ¸
from google.colab import drive
drive.mount('/content/drive')

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

# âœ… ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸° (íŒŒì¼ ê²½ë¡œ ìˆ˜ì •)
train_file_path = "/content/drive/MyDrive/train.csv"
test_file_path = "/content/drive/MyDrive/test.csv"

df_train = pd.read_csv(train_file_path)
df_test = pd.read_csv(test_file_path)

# âœ… ë ˆì´ë¸” ì¸ì½”ë”© (ë¬¸ì â†’ ìˆ«ì ë³€í™˜)
label_encoder = LabelEncoder()
df_train['label'] = label_encoder.fit_transform(df_train['label'])

# âœ… ë°ì´í„° ë¶„ë¦¬ (ID, label, í”½ì…€ ë°ì´í„°)
X_train = df_train.drop(['ID', 'label'], axis=1).to_numpy().astype('float32')
y_train = df_train['label'].to_numpy().astype('int')

X_test = df_test.drop(['ID'], axis=1).to_numpy().astype('float32')
test_ids = df_test['ID'].values

# âœ… ì •ê·œí™” (0~1ë¡œ ìŠ¤ì¼€ì¼ë§)
X_train /= 255.0
X_test /= 255.0

# âœ… ì°¨ì› ë³€í™˜ (CNN ì…ë ¥ í˜•íƒœë¡œ)
X_train = X_train.reshape(-1, 32, 32, 1)
X_test = X_test.reshape(-1, 32, 32, 1)

print(f"ğŸ”¹ ë³€í™˜ëœ X_train shape: {X_train.shape}")
print(f"ğŸ”¹ ë³€í™˜ëœ X_test shape: {X_test.shape}")

# âœ… í´ë˜ìŠ¤ë³„ ìƒ˜í”Œ ê°œìˆ˜ í™•ì¸ (ë°ì´í„° ë¶ˆê· í˜• ì²´í¬)
print("í´ë˜ìŠ¤ë³„ ìƒ˜í”Œ ê°œìˆ˜:", np.bincount(y_train))

# âœ… í›ˆë ¨ ë°ì´í„°ì™€ ê²€ì¦ ë°ì´í„° ë¶„ë¦¬ (90:10)
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(
    X_train, y_train, test_size=0.1, random_state=42, stratify=y_train
)

# âœ… ë°ì´í„° ì¦ê°• (ì ì ˆí•œ ë²”ìœ„ë¡œ ì¡°ì •)
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)
datagen.fit(X_train_split)

# âœ… CNN ëª¨ë¸ ë‹¨ìˆœí™” (í•„í„° ê°œìˆ˜ ì¤„ì„)
num_classes = len(label_encoder.classes_)

model = models.Sequential([
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32, 32, 1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(num_classes, activation='softmax')
])

# âœ… í•™ìŠµë¥  ì„¤ì • (0.001)
learning_rate = 0.001
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

# âœ… Early Stopping ì„¤ì •
callbacks = [
    EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
]

# âœ… ëª¨ë¸ ì»´íŒŒì¼
model.compile(optimizer=optimizer,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# âœ… ëª¨ë¸ í•™ìŠµ
history = model.fit(
    datagen.flow(X_train_split, y_train_split, batch_size=32),
    validation_data=(X_val_split, y_val_split),
    epochs=60,
    callbacks=callbacks
)

# âœ… í•™ìŠµ ê²°ê³¼ ì‹œê°í™” (ê³¼ì í•© ì—¬ë¶€ í™•ì¸)
plt.figure(figsize=(12, 4))

# ì •í™•ë„ ê·¸ë˜í”„
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training & Validation Accuracy')

# ì†ì‹¤ ê·¸ë˜í”„
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title('Training & Validation Loss')

plt.show()

# âœ… í…ŒìŠ¤íŠ¸ ë°ì´í„° ì˜ˆì¸¡
pred_probs = model.predict(X_test)
pred_labels = np.argmax(pred_probs, axis=1)

# âœ… ì œì¶œ íŒŒì¼ ìƒì„±
df_submit = pd.DataFrame({'ID': test_ids, 'label': label_encoder.inverse_transform(pred_labels)})
df_submit.to_csv('/content/drive/MyDrive/submission.csv', index=False)

print("ğŸ‰ ì œì¶œ íŒŒì¼ 'submission.csv' ìƒì„± ì™„ë£Œ! ğŸš€")
```
